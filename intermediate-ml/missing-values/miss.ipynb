### REMOVE MISSING VALUES

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
def socre_dataset(x_train, x_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(x_train,y_train)
    preds = model.predict(x_valid)
    return mean_absolute_error(y_valid,preds)

    
```

```python
cols_with_missing = [col for col in X_train.columns
                     if X_train[col].isnull().any()]
```

```python
redused_X_train = X_train.drop(cols_with_missing, axis=1)
reduced_X_valid = X_valid.drop(cols_with_missing, axis=1)
```

```python
print('MAE from reduced dataset:')
print(socre_dataset(redused_X_train,reduced_X_valid,y_train,y_valid))
```

### fill missing values with the mean of each column

```python
from sklearn.impute import SimpleImputer
```

```python
my_imputer = SimpleImputer(strategy='mean')

imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))

imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns
```

```python
print("MAE from imputed dataset:")
print(socre_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

```python
final_imputer = SimpleImputer(strategy='median')
final_X_train = pd.DataFrame(final_imputer.fit_transform(X_train))
final_X_valid = pd.DataFrame(final_imputer.transform(X_valid))
final_X_train.columns = X_train.columns
final_X_valid.columns = X_valid.columns
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(final_X_train, y_train)
pred_valid= model.predict(final_X_valid)
print("MAE from final dataset:")
print(mean_absolute_error(y_valid, pred_valid))
```

```python
final_X_test = pd.DataFrame(final_imputer.transform(x_test))
pred_test = model.predict(final_X_test)
```

```python
output = pd.DataFrame({'ID' :x_test.index,
                       'SalePrice' :pred_test})

output.to_csv('submission.csv', index=False)
```

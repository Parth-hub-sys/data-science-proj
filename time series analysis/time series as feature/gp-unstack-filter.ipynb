```python
import pandas as pd
import numpy as np

# Sample Data
data = {
    'date': pd.date_range('2017-01-01', periods=6, freq='D').tolist() * 2,
    'family': ['BEVERAGES']*6 + ['BREAD/BAKERY']*6,
    'sales': np.random.randint(10, 50, 12),
    'onpromotion': np.random.randint(0, 5, 12)
}

store_sales = pd.DataFrame(data)
store_sales.head()
```

```python
family_sales = (
    store_sales
    .groupby(['family', 'date'])
    .mean()
    .unstack('family')
    .loc['2017', ['sales', 'onpromotion']]
)
```

```python
family_sales.head()
```

```python
import matplotlib.pyplot as plt

family_sales['sales'].plot(title='Sales per Family (2017)', figsize=(10, 5))
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
family_sales['onpromotion'].plot(kind='bar', figsize=(10, 5), title="Daily Promotions")
```
